```python
# Import the libraries
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Load the data
texts = [
    "Data science helps companies make decisions",
    "Machine learning finds patterns in data",
    "Python is popular for data analysis",
    "Visualization makes complex data simple",
    "Artificial intelligence is the future",
    "Data analysis requires statistics knowledge",
    "Programming skills are important in technology",
    "Big data is growing rapidly",
    "Learning data science improves career opportunities",
    "Good visualization tells a story"
]
```

```python
# Vectorizer section
vectorizer = CountVectorizer(stop_words ='english')
X = vectorizer.fit_transform(texts)
```

```python
# PCA Section
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X.toarray())
```

```python
# Visualization section..
plt.figure(figsize=(6,5))

plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    color = 'salmon'
)

for i, t in enumerate(texts):
    plt.annotate(f'S{i}',(X_pca[i,0], X_pca[i,1]))

plt.title('PCA Visualization', fontsize = 15, fontweight ='bold')
plt.xlabel('PC1')
plt.ylabel('PC2')

plt.tight_layout()

# Siving this as a image
plt.savefig('pca_visualization.png', dpi=300, bbox_inches='tight')

plt.show()
```
