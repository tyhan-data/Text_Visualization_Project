```python
# Import the libraries
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# ✅ Task 1 — Convert Text to Numbers

```python
texts = [
    "Data science helps companies make decisions",
    "Machine learning finds patterns in data",
    "Python is popular for data analysis",
    "Visualization makes complex data simple",
    "Artificial intelligence is the future",
    "Data analysis requires statistics knowledge",
    "Programming skills are important in technology",
    "Big data is growing rapidly",
    "Learning data science improves career opportunities",
    "Good visualization tells a story"
]
```

```python
# Converting text in to Numbers
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(texts)

word_freq = dict(zip(vectorizer.get_feature_names_out(), X.toarray().sum(axis=0)))
```

# ✅ Task 2 — Find Top 10 Words

```python
# Separating into names and counts
names = vectorizer.get_feature_names_out()
counts = X.toarray().sum(axis = 0)
```

```python
# Now converting into dataframe
df = pd.DataFrame({
    'word':names,
    'count':counts
})
```

```python
# Now find out the top 10 words
top10 = df.sort_values(by='count', ascending=False).head(10)
top10.reset_index()
```

# ✅ Task 3 — Bar Chart Visualization ⭐

```python
# Now it is time to visualization
plt.figure(figsize=(6,4))

sns.barplot(
    x = top10['word'],
    y = top10['count'],
    data = top10,
    color='skyblue',
    edgecolor='black')

plt.title('Top 10 Words', color='darkred',family='serif',size=20, fontweight='bold')
plt.xlabel('Words', color='gray',family='serif',size=15)
plt.ylabel('Frequency', color='gray',family='serif',size=15)

plt.xticks(rotation = 45, fontsize = 12)
plt.yticks(fontsize = 12)

# Saving the image
plt.savefig( 'bar_chart.png', dpi = 300, bbox_inches ='tight')

plt.show()
```

# ✅ Task 4 — WordCloud Visualization ⭐⭐

```python
# Converting into WordCloud
wc = (
    WordCloud(height=400, width=900, background_color='white')
   .generate_from_frequencies(word_freq)
)
```

```python
# Now it is time to WordCloud Visualization
plt.figure(figsize=(10,5))

plt.imshow(wc, interpolation='bilinear')
plt.axis('off')

plt.title('Most Common Words In Dataset',
         fontsize = 16,
         fontweight ='bold')

plt.tight_layout()

# saving the WordCloud.
plt.savefig( 'WordCould_by_CountVectorizer.png', dpi = 300, bbox_inches = 'tight')

plt.show()
```

# ✅ Task 5 — TF-IDF Visualization (Important DS Skill)

```python
# Create a TF-IDF object
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(texts)

word_freq = dict(zip(
    vectorizer.get_feature_names_out(),
    X.toarray().sum(axis=0)
))
```

```python
# Now it's time to WordCloud Visualization.
wc = WordCloud(
    height = 400,
    width = 800,
    background_color = 'white'
).generate_from_frequencies(word_freq)

# Visualization section.
plt.figure(figsize=(12,5))

plt.imshow(wc, interpolation ='bilinear')
plt.axis('off')

plt.title('Most Important Word By TF-IDF Score',
         fontsize = 18,
         fontweight = 'bold',
         family = 'serif'
)

plt.tight_layout()

# saving the WordCloud.
plt.savefig( 'WordCloud_by_TF-IDF.png', dpi = 300, bbox_inches ='tight')

plt.show()
```
