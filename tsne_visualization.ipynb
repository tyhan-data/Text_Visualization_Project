```python
# First import the libraries
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# Load dataset
newsgroups = fetch_20newsgroups(subset ='train', categories =['sci.space'])
texts = newsgroups.data[:100]
```

```python
# Vectorizer section.
vectorizer = CountVectorizer(stop_words ='english', max_features=1000)
X = vectorizer.fit_transform(texts)
```

```python
# T-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X.toarray())
```

```python
# Now it's time to  Visualization 
plt.figure(figsize=(7,5))

plt.scatter( X_tsne[:,0], X_tsne[:,1])

# Label and title section
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('t-SNE Visualization of Text Data')

# Saving this as a image
plt.savefig('t-sne visualization', dpi=300, bbox_inches='tight')

plt.show()
```
